```python
mi_juego= list(range(1,101))
shuffle(mi_juego) 
numero_juego = mi_juego[19]
numero_juego
#mi_juego
```

```python
def adivinacion():
    intento = ''
    intento = int(input("Escoje un n√∫mero del 1 al 100"))
    while intento not in mi_juego:
        print("FUERA DE LIMITES")
        intento = int(input("Escoje un n√∫mero del 1 al 100"))
    return (intento)
```

```python
intento = adivinacion()
```

```python
lista_intentos = []
```

```python
while True:
    if len(lista_intentos) == 0:
        dif = intento - numero_juego
        if intento == numero_juego:
            print(f"Has acertado con {len(lista_intentos)} intentos")
            break
        elif abs(dif) <= 10 and dif != 0:
            print("CALIENTE")
            lista_intentos.append(intento)
        else:
            print("FR√çO")
            lista_intentos.append(intento)
    else:
        intento = adivinacion()
        if intento == numero_juego:
            print(f"Has acertado con {len(lista_intentos)} intentos")
            break
        elif abs(intento - numero_juego) < abs(lista_intentos[-1] - numero_juego):
            print("¬°M√ÅS CALIENTE!")
            lista_intentos.append(intento)
        else:
            print("¬°M√ÅS FR√çO!")
            lista_intentos.append(intento)
```
