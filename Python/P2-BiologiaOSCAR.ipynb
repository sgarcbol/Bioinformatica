#### 7.- Con la secuencia y la lista siguiente, escribe un algoritmo que cuente el número de veces que aparece un elemento de la lista en la secuencia. Pista: hacer un diccionario para almacenar los resultados puede ser buena idea

```python
dna = "AATGATGAACGAC" 
dinucleotides = ['AA','AT','AG','AC', 
                 'TA','TT','TG','TC', 
                 'GA','GT','GG','GC', 
                 'CA','CT','CG','CT'] 
```

```python
# Primero creamos un diccionario vacío para guardar los resultados, donde la clave será el dinucleotido y el valor el numero de veces que aparece en dna

counts = {}

# Ahora creamos un algoritmo que recorre cada elemento di de la lista y devuelve cuantas veces aparece la subcadena di en dna

for di in dinucleotides:
    counts[di] = dna.count(di)

# Mostramos los resultados

for k, v in counts.items():
    print(k, ":", v)         # Donde k es clave (key) y v el valor (value)
```

```python
# Hay que tener en cuenta que str.count cuenta ocurrencias no solapadas, pero en esta secuencia concreta las cuentas coinciden con las solapadas para la mayoría de casos. De todas formas creamos otro algoritmo para contar las solapadas y comparamos ambas:

dna = "AATGATGAACGAC" 

dinucleotides = [
    'AA','AT','AG','AC', 
    'TA','TT','TG','TC', 
    'GA','GT','GG','GC', 
    'CA','CT','CG','CT'
]

# 1) Eliminamos los términos duplicados y ordenamos alfabeticamente

targets = sorted(set(dinucleotides))

# 2) Cremaos diccionarios para los dos métodos

counts_solapado = {di: 0 for di in targets}
counts_no_solapado = {di: dna.count(di) for di in targets}  # usa str.count()

# 3) Creamos un bucle que recorre todas las posiciones posibles de una ventana de tamaño 2

for i in range(len(dna) - 1):        # Esto garantiza que siempre tenga 2 caracteres
    di = dna[i:i+2]                  # Este rebanado extrae el dinucleótido que empieza en i
    if di in counts_solapado:        # Si el dinucleótido extraido está en nuestros objetivos (targets) incrementa su contador, mientras que el if evita contar dinucleótidos que no estén en la lista
        counts_solapado[di] += 1

# 4) Mostrar comparación
print("Dinucleótido | Solapado | No solapado")
print("-------------------------------------")
for di in targets:
    print(f"{di:11} | {counts_solapado[di]:8} | {counts_no_solapado[di]:11}")

# Como vemos que el resultado es el mismo independientemente del método utilizado, empleamos la versión mas facil del algoritmo para las siguientes preguntas
```

##### ¿Cuántos counts tiene TC?

```python
# Incluso usando un conteo solapado TC aparece 0 veces ya que no está en la secuencia AATGATGAACGAC, de todas formas lo comprobamos con código:

dna = "AATGATGAACGAC"

# Inicializar contador
count_TC = 0

# Recorre la secuencia con ventana de 2
for i in range(len(dna) - 1):
    di = dna[i:i+2]      # extrae un dinucleótido
    if di == "TC":       # si es "TC"
        count_TC += 1    # suma al contador

print("Ocurrencias de TC:", count_TC)
```

##### ¿Puedes imprimir los dinuleótidos que tengan counts igual a 2?

```python
# Para ello debemos imprimir solo los que tienen count == 2, vamos a comparar el numero en un conteo solapado y no solapado:

dna = "AATGATGAACGAC"

dinucleotides = [
    'AA','AT','AG','AC', 
    'TA','TT','TG','TC', 
    'GA','GT','GG','GC', 
    'CA','CT','CG','CT'
]

# Eliminamos duplicados y ordenamos
targets = sorted(set(dinucleotides))

# Hacemos un conteo solapado (como el resultado es el mismo en un conteo solapado que en uno no solapamos podemos simplificar el algoritmo
counts = {di: 0 for di in targets}
for i in range(len(dna) - 1):
    di = dna[i:i+2]
    if di in counts:
        counts[di] += 1

# Mostrar solo los que tienen count == 2
print("Dinucleótidos con count = 2:")
for di, val in counts.items():
    if val == 2:
        print(di, ":", val)
```
