#### 7.- Realiza una función que mediante la introducción de una secuencia de ADN devuelva lo siguiente: (3pto)
1. Todos los ORFs posibles que puedan surgir de la secuencia.
2. Selecciona los 6 ORFs más largos e indica si corresponden a la strand positiva o negativa.
3. Devuelve como mejor predicción, el más largo de todos.
4. Calcula el porcentaje de aminoácidos aromáticos de esa proteína elegida y su longitud relativa a la secuencia original de ADN.

```python
# Primero defino la cadena complementaria del adn y uso la función upper para poner todo en mayúsculas 
def complementaria(dna):
    dna = dna.upper()
    comp = ""
    for nucleotido in adn:
        if nucleotido == "A":
            comp = comp + "T"
        elif nucleotido == "T":
            comp = comp + "A"
        elif nucleotido == "C":
            comp = comp + "G"
        elif nucleotido == "G":
            comp=  comp + "C"
    # Para devolver la cadena complementaria inversa
    return comp[::-1]
# Ahora el siguiente paso es encontrar los orfs teniendo en cuenta los codones de stop 
def encontrar_orfs(dna):
    dna = dna.upper().replace(" ", "").replace("\n", "")
    stop_codones = ["TAA", "TAG", "TGA"]
# Meto mis orfs en una lista 
    orfs = []

    for frame in range(3):  # Hay tres marcos de lectura posibles
        i = frame
        while i < len(dna) - 2:
            codon = dna[i:i+3]
            if codon == "ATG":  # es el codón de inicio
                for j in range(i+3, len(dna)-2, 3):
                    codon_stop = dna[j:j+3]
                    if codon_stop in stop_codones:
                        orf = dna[i:j+3]
                        orfs.append(orf)
                        break  # Si encuentra un stop que guarde el orf con la función .append
            i += 3
    return orfs # Para devolver los orfs
# Me creo mi cadena complementaria a la cadena de la secuencia_final 
dna_comp = complementaria(secuencia_final)
#print(dna_comp)
# Tengo que buscar los orfs tanto en la hebra positiva como negativa
listado_orf_positivo = encontrar_orfs(secuencia_final)
listado_orf_negativo = encontrar_orfs(dna_comp)
# Me genero un diccionario que guarde todos los orf
dict_total_orf = {}
count_pos = 0
count_neg = 0
# Recorro todos los ORFs de la hebra positiva y a cada uno le asigno un nombre con un número consecutivo y "+"
for orf in listado_orf_positivo:
    nombre_orf = "ORF" + str(count_pos) + "+"
    count_pos += 1
    dict_total_orf[nombre_orf] = orf # Lo guardo en el diccionario
# Lo mismo con la hebra negativa
for orf in listado_orf_negativo:
    nombre_orf = "ORF" + str(count_neg) + "-"
    count_neg += 1
    dict_total_orf[nombre_orf] = orf
    
#dict_total_orf
```
