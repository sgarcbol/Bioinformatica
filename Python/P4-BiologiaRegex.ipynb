##### 2.- Crea dos archivos. Uno almacena los accession que empiezen por 'a' y el otro el resto

```python
accs = ['ab56', 'bh84', 'hv76', 'ay93', 'ap97', 'bd72']
```

```python
lista_a=[]
lista_resto=[]
for accesion in accs:
    if accesion.startswith('a'):
        lista_a.append(accesion)
    else:
        lista_resto.append(accesion)
```

```python
archivo_accs= open('accs.txt','w+')
for a in lista_a:
    archivo_accs.write(f"{a}\n")
archivo_resto= open('resto.txt','w+')
for r in lista_resto:
    archivo_resto.write(f"{r}\n")
```
