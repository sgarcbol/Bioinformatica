#### Escribe un programa que coja un nÃºmero entero del 1 al 100, y los jugadores tienen que tratar de adivinar el nÃºmero.

#### Las reglas son las siguientes:

1. Si un jugador escoje un nÃºmero menor que 1 o mayor que 100, el programa dice "FUERA DE LÃMITES"
2. En el primer turno del jugador, si estÃ¡ en un rango de 10 del nÃºmero dice "Â¡CALIENTE!". Si estÃ¡ fuera de ese rango dice "Â¡FRIO!"
3. En todas las rondas siguientes, si el jugador se acerca mÃ¡s que la vez anterior, dice "Â¡MAS CALIENTE!". Si estÃ¡ mÃ¡s lejos que la anterior vez dice "Â¡MAS FRIO!"
4. Cuando el jugador acierte el nÃºmero, les dirÃ¡ que han acertado y cuÃ¡ntos intentos les ha llevado





##### PISTAS
1. Trata obtener un nÃºmero aleatorio entre el 1 y el 100
2. Haz una introducciÃ³n de tu juego y explica las reglas
3. Crea una lista que almacene los intentos. 
4. Haz un bucle que evalÃºe un nÃºmero. Prueba que funcione primero. While podrÃ­a ir bien
5. Modifica ese bucle para que compare nuestro nÃºmero con el intento. Si se acierta usa break, si no puedes usar continue y tienes que seguir preguntando al jugador por nuevos intentos

```python
# Primero importamos la librerÃ­a "random" para generar nÃºmeros aleatorios

import random

# Empezamos definiendo una funciÃ³n auxiliar "extra" que detecta si un intento ya ha sido propuesto con anterioridad

def intento_repetido(intento, intentos):                           # Comprobamos si el nÃºmero 'intento' estÃ¡ ya dentro de la lista 'intentos'
    if intento in intentos:
        print("Â¡Ya has adivinado ese nÃºmero y no era correcto!")   # Si estÃ¡ repetido, mostramos el mensaje solicitado
        return True                                                # Devolvemos True indicando que el intento es repetido
    return False                                                   # Si no estÃ¡ repetido, devolvemos False
    
# Ahora si, definimos la funciÃ³n principal del juego:

def juego_caliente_frio():
    
    # AÃ±adimos un mensaje de bienvenida y explicamos las reglas
    
    print("ğŸ® Bienvenido a 'Caliente o FrÃ­o'")
    print("Reglas:")
    print("1. Piensa en un nÃºmero entre 1 y 100.")
    print("2. Si el nÃºmero estÃ¡ fuera del rango (1-100), verÃ¡s 'FUERA DE LÃMITES'.")
    print("3. En el primer intento:")
    print("   - Si tu nÃºmero estÃ¡ dentro de 10 del nÃºmero secreto â 'Â¡CALIENTE!'")
    print("   - Si estÃ¡ fuera de ese rango â 'Â¡FRÃO!'")
    print("4. En intentos siguientes:")
    print("   - Si estÃ¡s mÃ¡s cerca que antes â 'Â¡MÃS CALIENTE!'")
    print("   - Si estÃ¡s mÃ¡s lejos â 'Â¡MÃS FRÃO!'")
    print("5. El juego termina cuando adivines el nÃºmero.\n")

    # Genera un nÃºmero aleatorio entre 1 y 100 que el jugador no podrÃ¡ ver
    
    numero_secreto = random.randint(1, 100)

    # Creamos una lista vacÃ­a para guardar los intentos del jugador
    
    intentos = []

    # Creamos un bucle infinito para seguir pidiendo intentos hasta que se acierte
    
    while True:
        try:
            intento = int(input("ğŸ‘‰ Adivina un nÃºmero entre 1 y 100: "))     # Pedimos al usuario que ingrese un nÃºmero
        except ValueError:
            print("âš ï¸ Debes escribir un nÃºmero entero.")                     # Si el jugador escribe algo que no sea nÃºmero, mostramos este aviso
            continue                                                         # Volvemos al inicio del bucle

        # Verificamos si el nÃºmero estÃ¡ dentro de los lÃ­mites permitidos
        
        if intento < 1 or intento > 100:
            print("FUERA DE LÃMITES")
            continue  # Volvemos a pedir nÃºmero

        # Guardamos el intento en la lista de intentos
        
        intentos.append(intento)

        # Caso: el jugador adivina el nÃºmero
        
        if intento == numero_secreto:
            print(f"ğŸ‰ Â¡Has acertado! El nÃºmero era {numero_secreto}.")
            print(f"Te ha llevado {len(intentos)} intentos.")  # Mostramos cuÃ¡ntos intentos usÃ³
            break  # Salimos del bucle (fin del juego)

        # Caso: primer intento (se compara solo con el nÃºmero secreto)
        
        if len(intentos) == 1:
            
            # Si la diferencia con el nÃºmero secreto es menor o igual a 10 â†’ Â¡CALIENTE!
            
            if abs(numero_secreto - intento) <= 10:
                print("Â¡CALIENTE!")
            else:
                
                # Si estÃ¡ a mÃ¡s de 10 â†’ Â¡FRÃO!
                
                print("Â¡FRÃO!")
        else:
            
            # Caso: intentos siguientes, comparamos con el anterior
            # Si este intento estÃ¡ mÃ¡s cerca del secreto que el anterior â†’ Â¡MÃS CALIENTE!
            
            if abs(numero_secreto - intento) < abs(numero_secreto - intentos[-2]):
                print("Â¡MÃS CALIENTE!")
            else:
                
                # Si este intento estÃ¡ mÃ¡s lejos que el anterior â†’ Â¡MÃS FRÃO!"
                
                print("Â¡MÃS FRÃO!")

# Punto de entrada del programa: ejecuta el juego si se lanza directamente este archivo

if __name__ == "__main__":
    juego_caliente_frio()
```
