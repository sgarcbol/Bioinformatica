1. Todos los ORFs posibles que puedan surgir de la secuencia.

```python
# ajusto la secuencia para que no haya errores por caracteres inesperados otra vez
# (son mi peor enemigo)
secuencia = secuencia.upper()
secuencia = secuencia.replace("\n", "")
secuencia = secuencia.replace("\r", "")
secuencia = secuencia.replace(" ", "")

print("Se ha ajustado la secuencia. Longitud final:", len(secuencia))
print("ATG en la secuencia (despues del ajuste):", secuencia.count("ATG"))

# funcion para sacar la hebra negativa definiendola como revcomp (reversa complementaria)
def revcomp(seq):
    cadena = ""
    for nucleotido in seq:
        if nucleotido == "A":
            cadena = "T" + cadena
        elif nucleotido == "T":
            cadena = "A" + cadena
        elif nucleotido == "C":
            cadena = "G" + cadena
        elif nucleotido == "G":
            cadena = "C" + cadena
        else:
            cadena = "N" + cadena
    return cadena

codones_stop = ["TAA", "TAG", "TGA"]
# genero una lista donde guardo todos los ORFs
todos_los_orfs = []
# busco los ORFs en la hebra positiva
for i in range(len(secuencia)-2): # -2 para ajustar más y evitar cortes fuera de rango
    if secuencia[i:i+3] == "ATG":
        for j in range(i+3, len(secuencia)-2, 3):
            codon = secuencia[j:j+3]
            if codon in codones_stop:
                orf_encontrado = secuencia[i:j+3]
                todos_los_orfs.append(orf_encontrado)
                break

# ahora se hace lo mismo pero para buscar en la hebla complementaria 
negativa = revcomp(secuencia)
for i in range(len(negativa)-2):
    if negativa[i:i+3] == "ATG":
        for j in range(i+3, len(negativa)-2, 3):
            codon = negativa[j:j+3]
            if codon in codones_stop:
                orf_encontrado = negativa[i:j+3]
                todos_los_orfs.append(orf_encontrado + "_rev")
                break

print("ORFs encontrados (total):", len(todos_los_orfs))
for encontrar, orf in enumerate(todos_los_orfs[:10]): # lo he limitado a 10 para que no salgan los 359 ORFs 
    print("ORF", encontrar+1, "longitud (bases):", len(orf))

# sacamos los 6 ORFs más largos (no dice exactamente cuales son porque antes solo había impreso los primeros 10)
# primero ordenamos todos los ORFs por su longitud 
orfs_ordenados = sorted(todos_los_orfs, key=len, reverse=True)
# selecciónamos ahora los 6 más largos
top6_orfs = orfs_ordenados[:6]
print("\nLos 6 ORFs más largos encontrados son:")

for i, orf in enumerate(top6_orfs):
# Compruebo si el ORF venía de la hebra negativa dándole el buscador de _rev 
# (en realidad el problema que daba es que según yo era imposible que todas fueran de la hebra positiva)
# (pero lo dejo, que queda más bonito)
    if "_rev" in orf:
        hebra = "negativa"
    else:
        hebra = "positiva"
    
# Quito el "_rev" para medir bien su longitud porque no para de darme problemas
    longitud_real = len(orf.replace("_rev", ""))
    
    print("ORF", i+1, "longitud:", longitud_real, "bases", "- Hebra:", hebra)
```

```python
orf_mas_largo = top6_orfs[0].replace("_rev","") # le quito el rev_ porque da problemas a la hora de buscar
print("\nEl ORF más largo tiene:", len(orf_mas_largo), "bases")

# defino la función para traducir ADN a proteína y le enseño toda la equivalencia del código genético
def traducir(seq):
    tabla_codones = {'ATA':'I', 'ATC':'I', 'ATT':'I', 'ATG':'M', 'ACA':'T', 'ACC':'T', 'ACG':'T', 'ACT':'T', 'AAC':'N', 
                     'AAT':'N', 'AAA':'K', 'AAG':'K', 'AGC':'S', 'AGT':'S', 'AGA':'R', 'AGG':'R', 'CTA':'L', 'CTC':'L', 
                     'CTG':'L', 'CTT':'L', 'CCA':'P', 'CCC':'P', 'CCG':'P', 'CCT':'P', 'CAC':'H', 'CAT':'H', 'CAA':'Q', 
                     'CAG':'Q', 'CGA':'R', 'CGC':'R', 'CGG':'R', 'CGT':'R', 'GTA':'V', 'GTC':'V', 'GTG':'V', 'GTT':'V',
                     'GCA':'A', 'GCC':'A', 'GCG':'A', 'GCT':'A', 'GAC':'D', 'GAT':'D', 'GAA':'E', 'GAG':'E', 'GGA':'G', 
                     'GGC':'G', 'GGG':'G', 'GGT':'G', 'TCA':'S', 'TCC':'S', 'TCG':'S', 'TCT':'S', 'TTC':'F', 'TTT':'F', 
                     'TTA':'L', 'TTG':'L', 'TAC':'Y', 'TAT':'Y', 'TAA':'*', 'TAG':'*', 'TGC':'C', 'TGT':'C', 'TGA':'*', 
                     'TGG':'W', }
    proteina = ""
    for i in range(0, len(seq)-2, 3): # le enseño que tiene que contar de 3 en 3, del tamaño de los codones
        codon = seq[i:i+3]
        if codon in tabla_codones:
            aa = tabla_codones[codon]
            if aa == "*":
                break
            proteina = proteina + aa
    return proteina

# traduzco el ORF más largo
proteina = traducir(orf_mas_largo)
print("Longitud de la proteína:", len(proteina), "aminoácidos")

# calculo porcentaje de aminoácidos aromáticos (F, Y, W: fenilalanina, tirosina y triptófano)
aromaticos = proteina.count("F") + proteina.count("Y") + proteina.count("W")
porcentaje_aromaticos = (aromaticos / len(proteina)) * 100 # para poder expresarlo en forma de porcentaje
print("Porcentaje de aminoácidos aromáticos:", round(porcentaje_aromaticos, 2), "%")

# longitud relativa respecto al ADN original
relativa = (len(orf_mas_largo) / len(secuencia)) * 100
print("Longitud relativa del ORF más largo respecto al ADN:", round(relativa, 2), "%")
```
