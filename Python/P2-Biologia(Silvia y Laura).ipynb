#### 7.- Con la secuencia y la lista siguiente, escribe un algoritmo que cuente el número de veces que aparece un elemento de la lista en la secuencia. Pista: hacer un diccionario para almacenar los resultados puede ser buena idea

```python
dna = "AATGATGAACGAC" 
dinucleotides = ['AA','AT','AG','AC', 
                 'TA','TT','TG','TC', 
                 'GA','GT','GG','GC', 
                 'CA','CT','CG','CT'] 
```

```python
#Creamos un diccionario vacío para posteriormente añadir la relación entre los dinucleótidos y sus valores mediante la función .count() 
diccionario_vacio= {}
for dnt in dinucleotides: 
    if dnt in dna:
        a=dna.count(dnt)
        diccionario_vacio[dnt] = a
diccionario_vacio
```

##### ¿Cuántos counts tiene TC?

```python
#Aplicamos .count() para contar cuántas veces se repite TC
dna.count('TC')
```

##### ¿Puedes imprimir los dinuleótidos que tengan counts igual a 2?

```python
#Creamos un bucle for para imprimir los dinucleótidos que tengan counts igual a 2
for key in diccionario_vacio.keys():
    if diccionario_vacio[key] == 2:
        print(key)
```
